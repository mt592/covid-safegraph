```python
import pandas as pd
import ast
pd.set_option('display.max_columns', None) 
```

Weekly Pattern Data

```python
path = "~/Desktop/covid/data/weekly_patterns/v1/main-file/"
```

```python
iter_csv = pd.read_csv(path+"2020-04-12-weekly-patterns.csv.gz",
                       escapechar='\\',
                       compression='gzip', 
                       iterator=True, 
                       chunksize=1000)

df = pd.concat([chunk[chunk['city'] == 'Austin'] for chunk in iter_csv])
df = pd.DataFrame(df)
```

```python
print(df.shape)
df = df.reset_index()
```

```python
df.head(10)
```

```python
df.visits_by_day = df.visits_by_day.apply(ast.literal_eval)
df = df.explode('visits_by_day')
```

```python
print(df.shape)
df.head(10)
```

```python
day_expand = df.apply(lambda row: pd.Series(row))
#cbg_expand = pd.concat([df, cbg_expand], axis=1)
```

Social Distancing Data

```python
path = "~/covid/data/social-distancing/2020/04/20/"

iter_csv = pd.read_csv(path+"2020-04-20-social-distancing.csv.gz",
                       escapechar='\\',
                       compression='gzip', 
                       iterator=True, 
                       chunksize=1000)

sd = pd.concat([chunk[chunk['origin_census_block_group'].astype('str').str.contains('48453', regex=False)] for chunk in iter_csv])
sd = pd.DataFrame(sd)
```
